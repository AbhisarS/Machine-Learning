```python
#Program to show the usage of logistic regression.
#Importing necessary libraries.
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
import numpy  as np
import pandas as pd

#Understanding the dataset.
wine=load_wine()
print("Description of dataset:", wine.DESCR)
print("Dimension of independent variables:", wine.data.shape)
print("Dimension of dependent variables:", wine.target.shape)
print("Name of variables:\n", wine.feature_names)
print("Keys in the dataset:\n", wine.keys())

#Convert it into data frame.
datadf = pd.DataFrame(wine.data)
targetdf = pd.DataFrame(wine.target)

#Determining missing values.
print("Missing Values:\n", datadf.isnull().sum())

#Generating the same dataset.
np.random.seed(1000)

#Creating training, test dataset for dependent and independent variables.
x_trg, x_test, y_trg, y_test = train_test_split(datadf, targetdf, random_state=0)
print("Dimension of training dataset:", x_trg.shape)
print("Dimension of test dataset:", x_test.shape)

#Developing the model
#Creating a logistic regression model.
winemodel = LogisticRegression()

#Fitting the model.
winemodel.fit(x_trg, y_trg)

#Determining score of training and test dataset.
print("Training set score:", winemodel.score(x_trg, y_trg))
print("Test set score:", winemodel.score(x_test, y_test))

#Predicting and determining accuracy of the model
pred = winemodel.predict(x_test)

#Determining accuracy nusing confusion matrix
results = confusion_matrix(y_test, pred)
print("Confusion Matrix:\n", results)
print("Accuracy:", accuracy_score(y_test, pred))
```

```python
# Logistic Regression with Data Processing on Wine Dataset

# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Set plot style
sns.set(style="whitegrid")

# Load the dataset
wine = load_wine()
X = pd.DataFrame(wine.data, columns=wine.feature_names)
y = pd.Series(wine.target, name='target')

print("Dataset shape:", X.shape)
print("First 3 rows of dataset:\n", X.head(3))

# --------------------------- Visualization -----------------------------

print("\n---------------- DISTRIBUTION PLOTS ----------------")
plt.figure(figsize=(18, 12))
for i, col in enumerate(['alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash',
                         'magnesium', 'total_phenols', 'flavanoids',
                         'nonflavanoid_phenols', 'proanthocyanins']):
    plt.subplot(3, 3, i+1)
    sns.histplot(X[col], bins=15, kde=True)
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()

print("\n---------------- HISTOGRAMS ----------------")
plt.figure(figsize=(15, 5))
for i, col in enumerate(['color_intensity', 'hue', 'proline']):
    plt.subplot(1, 3, i+1)
    X[col].hist(bins=10)
    plt.title(f"Histogram of {col}")
plt.tight_layout()
plt.show()

print("\n---------------- HEATMAP ----------------")
plt.figure(figsize=(12, 10))
sns.heatmap(X.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Feature Correlations')
plt.show()

print("\n---------------- SCATTER PLOT ----------------")
sns.scatterplot(x='color_intensity', y='hue', data=X)
plt.title("Scatter plot: Color Intensity vs Hue")
plt.xlabel("Color Intensity")
plt.ylabel("Hue")
plt.show()

print("\n---------------- SCATTER PAIR PLOT ----------------")
sns.pairplot(X[['alcalinity_of_ash', 'total_phenols', 'flavanoids',
                'nonflavanoid_phenols', 'proanthocyanins']],
             diag_kind='kde')
plt.suptitle("Pair Plot of Selected Features", y=1.02)
plt.show()

print("\n---------------- REGRESSION PAIR PLOT ----------------")
sns.pairplot(X[['alcohol', 'magnesium', 'ash', 'malic_acid', 'proline']],
             kind='reg', diag_kind='hist')
plt.suptitle("Regression Pair Plot of Selected Features", y=1.02)
plt.show()

# ------------------------ Data Cleaning & Preprocessing ------------------------

# Check for missing values
print("\nMissing values:\n", X.isnull().sum())

# Normality checks
print("\nSkewness:\n", pd.Series(stats.skew(X)))
print("\nKurtosis:\n", pd.Series(stats.kurtosis(X)))

# Outlier detection using Z-score
z_scores = np.abs(stats.zscore(X))
outlier_indices = np.where(z_scores > 3)[0]
print("\nNumber of outliers detected:", len(outlier_indices))

# Remove outliers
X_clean = X[(z_scores < 3).all(axis=1)]
y_clean = y.drop(index=outlier_indices).reset_index(drop=True)
X_clean = X_clean.reset_index(drop=True)

print("Dataset shape after removing outliers:", X_clean.shape)

# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_clean)

# ------------------------ Model Development ------------------------

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_clean, test_size=0.25, random_state=0)

print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)

# Model training
model = LogisticRegression(max_iter=200, multi_class='auto')
model.fit(X_train, y_train)

# Evaluation
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)
y_pred = model.predict(X_test)

print("\nTraining Accuracy:", train_score)
print("Test Accuracy:", test_score)

# Confusion Matrix and Classification Report
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
