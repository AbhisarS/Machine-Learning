```python
# =============================
# Colab Notebook: Imbalanced Learning Playground (UCI Default of Credit Card Clients)
# Includes: rich visualizations, multiple resampling methods (SMOTE/ADASYN/Borderline/KMeans-SMOTE),
# a custom interpolation method (LCI), and model comparisons (LogReg / RandomForest / XGBoost)
# =============================

#  0) Install dependencies (Colab)
!pip install -q ucimlrepo imbalanced-learn xgboost seaborn matplotlib pandas scikit-learn

#  1) Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from ucimlrepo import fetch_ucirepo
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    roc_curve,
    precision_recall_curve,
    average_precision_score
)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from imblearn.over_sampling import SMOTE, ADASYN, BorderlineSMOTE, KMeansSMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

#  2) Plotting helpers
sns.set(context="notebook", style="whitegrid")

def plot_class_distribution(y, title="Class Distribution"):
    plt.figure(figsize=(5,3))
    ax = sns.countplot(x=pd.Series(y, name='Class'))
    plt.title(title)
    for p in ax.patches:
        ax.annotate(f"{int(p.get_height())}", (p.get_x()+p.get_width()/2., p.get_height()),
                    ha='center', va='bottom', fontsize=9)
    plt.show()

def plot_corr_heatmap(df, title="Correlation Heatmap"):
    plt.figure(figsize=(10,7))
    sns.heatmap(df.corr(numeric_only=True), annot=False, cmap="coolwarm", center=0)
    plt.title(title)
    plt.show()

def plot_pca(X, y, title="PCA (2D)"):
    pca = PCA(n_components=2, random_state=42)
    Xp = pca.fit_transform(X)
    plt.figure(figsize=(6,5))
    sns.scatterplot(x=Xp[:,0], y=Xp[:,1], hue=y, s=18, alpha=0.8)
    plt.title(title)
    plt.legend(title='Class', loc='best')
    plt.show()

def plot_tsne(X, y, title="t-SNE (2D)"):
    ts = TSNE(n_components=2, perplexity=30, learning_rate='auto', init='pca', random_state=42)
    Xt = ts.fit_transform(X)
    plt.figure(figsize=(6,5))
    sns.scatterplot(x=Xt[:,0], y=Xt[:,1], hue=y, s=15, alpha=0.8)
    plt.title(title)
    plt.legend(title='Class', loc='best')
    plt.show()

def plot_confusion(cm, title="Confusion Matrix"):
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

def plot_roc_curves(results, y_test, title_prefix="ROC"):
    plt.figure(figsize=(6,5))
    for name, (clf, X_test_scaled) in results.items():
        if hasattr(clf, "predict_proba"):
            y_score = clf.predict_proba(X_test_scaled)[:,1]
        else:
            # fallback to decision_function if available
            y_score = clf.decision_function(X_test_scaled)
        fpr, tpr, _ = roc_curve(y_test, y_score)
        auc = roc_auc_score(y_test, y_score)
        plt.plot(fpr, tpr, label=f"{name} (AUC={auc:.3f})")
    plt.plot([0,1],[0,1], 'k--', alpha=0.5)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f"{title_prefix} Curves")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
    plt.show()

def plot_pr_curves(results, y_test, title_prefix="Precision-Recall"):
    plt.figure(figsize=(6,5))
    for name, (clf, X_test_scaled) in results.items():
        if hasattr(clf, "predict_proba"):
            y_score = clf.predict_proba(X_test_scaled)[:,1]
        else:
            y_score = clf.decision_function(X_test_scaled)
        precision, recall, _ = precision_recall_curve(y_test, y_score)
        ap = average_precision_score(y_test, y_score)
        plt.plot(recall, precision, label=f"{name} (AP={ap:.3f})")
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(f"{title_prefix} Curves")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
    plt.show()

#  3) Custom interpolation method (NEW): Local Cluster Interpolation (LCI)
# Idea: For each minority point, find k nearest minority neighbors and create synthetic points by
#       interpolating with a random neighbor + small Gaussian noise scaled by feature std.
from sklearn.neighbors import NearestNeighbors

def lci_resample(X, y, ratio=1.0, k_neighbors=5, noise_scale=0.05, random_state=42):
    rng = np.random.default_rng(random_state)
    X = np.asarray(X)
    y = np.asarray(y).astype(int)
    X_min = X[y==1]
    X_maj = X[y==0]

    n_min, n_maj = len(X_min), len(X_maj)
    target_n_min = int(max(n_min, ratio * n_maj))
    n_new = max(0, target_n_min - n_min)
    if n_new == 0:
        return X.copy(), y.copy()

    # Fit NN on minority only
    nn = NearestNeighbors(n_neighbors=min(k_neighbors+1, len(X_min)))
    nn.fit(X_min)
    nbrs = nn.kneighbors(X_min, return_distance=False)

    # Feature-wise std to scale noise
    feat_std = X_min.std(axis=0, ddof=1)
    feat_std[feat_std==0] = 1.0

    synthetic = []
    for _ in range(n_new):
        i = rng.integers(0, len(X_min))
        # choose a random neighbor (skip self at index 0)
        idx = rng.choice(nbrs[i][1:]) if len(nbrs[i]) > 1 else i
        lam = rng.random()  # interpolation factor in [0,1]
        x_new = X_min[i] * lam + X_min[idx] * (1 - lam)
        # add small noise
        noise = rng.normal(0, noise_scale, size=x_new.shape) * feat_std
        x_new = x_new + noise
        synthetic.append(x_new)

    X_syn = np.vstack([X, np.array(synthetic)])
    y_syn = np.hstack([y, np.ones(n_new, dtype=int)])
    return X_syn, y_syn

#  4) Load UCI dataset (Default of Credit Card Clients)
# id=350 via ucimlrepo
uci = fetch_ucirepo(id=350)
X_df = uci.data.features.copy()
y_series = uci.data.targets.copy().squeeze()  # ensure 1D Series

# The target column is named 'default payment next month' in UCI sheet
if isinstance(y_series, pd.DataFrame):
    y_series = y_series.iloc[:,0]

print("Shapes:", X_df.shape, y_series.shape)
print("Class ratio (original):\n", y_series.value_counts(normalize=True))
plot_class_distribution(y_series, title="Original Class Distribution")

# Optional correlation heatmap (raw features)
plot_corr_heatmap(pd.concat([X_df, y_series.rename('target')], axis=1), title="Correlation (Raw Features)")

#  5) Train/Test split + scaling
X_train_df, X_test_df, y_train, y_test = train_test_split(
    X_df, y_series, test_size=0.2, random_state=42, stratify=y_series
)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train_df)
X_test = scaler.transform(X_test_df)

# Visualize original train space
plot_pca(X_train, y_train, title="PCA (Train)  Original")
# t-SNE can be slow; keep enabled, adjust perplexity if needed
plot_tsne(X_train, y_train, title="t-SNE (Train)  Original")

#  6) Resampling options
resamplers = {
    "NONE": None,
    "SMOTE": SMOTE(random_state=42),
    "ADASYN": ADASYN(random_state=42),
    "BORDERLINE": BorderlineSMOTE(random_state=42),
    "KMEANS_SMOTE": KMeansSMOTE(random_state=42),
    "LCI_CUSTOM": "LCI"  # our custom interpolation method
}

#  7) Models to compare
models = {
    "LogReg": LogisticRegression(max_iter=2000, n_jobs=None, class_weight=None),
    "RandomForest": RandomForestClassifier(n_estimators=300, max_depth=None, random_state=42, n_jobs=-1, class_weight=None),
    "XGBoost": XGBClassifier(
        n_estimators=500, max_depth=5, learning_rate=0.05, subsample=0.9, colsample_bytree=0.9,
        reg_lambda=1.0, random_state=42, eval_metric='logloss', tree_method='hist'
    )
}

#  8) Train/Evaluate across resamplers and models
summary_rows = []
roc_plot_inputs = {}
pr_plot_inputs = {}

for rname, r in resamplers.items():
    # Resample only on TRAIN
    if r is None:  # no resampling
        Xtr_bal, ytr_bal = X_train.copy(), y_train.values.copy()
    elif r == "LCI":
        Xtr_bal, ytr_bal = lci_resample(X_train, y_train.values, ratio=1.0, k_neighbors=5, noise_scale=0.03, random_state=42)
    else:
        Xtr_bal, ytr_bal = r.fit_resample(X_train, y_train)

    print(f"\n=== Resampler: {rname} ===")
    plot_class_distribution(ytr_bal, title=f"Class Distribution  {rname}")

    # Visualize balanced training space
    plot_pca(Xtr_bal, ytr_bal, title=f"PCA (Train)  {rname}")

    # Optional t-SNE; comment out if slow
    # plot_tsne(Xtr_bal, ytr_bal, title=f"t-SNE (Train)  {rname}")

    for mname, clf in models.items():
        print(f"\n[Model: {mname} | Resampler: {rname}]")
        # Fit on balanced train
        clf.fit(Xtr_bal, ytr_bal)
        # Evaluate on untouched test set
        y_pred = clf.predict(X_test)
        if hasattr(clf, "predict_proba"):
            y_score = clf.predict_proba(X_test)[:,1]
        else:
            y_score = clf.decision_function(X_test)

        cm = confusion_matrix(y_test, y_pred)
        auc = roc_auc_score(y_test, y_score)
        ap = average_precision_score(y_test, y_score)

        print("AUC:", round(auc, 4), "| AP:", round(ap, 4))
        print(classification_report(y_test, y_pred))
        plot_confusion(cm, title=f"Confusion  {mname} ({rname})")

        key = f"{mname} ({rname})"
        roc_plot_inputs[key] = (clf, X_test)
        pr_plot_inputs[key] = (clf, X_test)

# Combined ROC & PR plots across all runs
plot_roc_curves(roc_plot_inputs, y_test, title_prefix="ROC")
plot_pr_curves(pr_plot_inputs, y_test, title_prefix="Precision-Recall")

#  9) Feature importance visualizations (tree models + LR coefficients)
# RandomForest feature importances
rf = RandomForestClassifier(n_estimators=400, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
rf_imp = pd.Series(rf.feature_importances_, index=X_train_df.columns).sort_values(ascending=False).head(20)
plt.figure(figsize=(7,5))
rf_imp[::-1].plot(kind='barh')
plt.title("RandomForest  Top 20 Feature Importances")
plt.show()

# XGBoost feature importances
xgb = XGBClassifier(n_estimators=600, max_depth=5, learning_rate=0.05, subsample=0.9, colsample_bytree=0.9,
                    reg_lambda=1.0, random_state=42, eval_metric='logloss', tree_method='hist')
xgb.fit(X_train, y_train)
xgb_imp = pd.Series(xgb.feature_importances_, index=X_train_df.columns).sort_values(ascending=False).head(20)
plt.figure(figsize=(7,5))
xgb_imp[::-1].plot(kind='barh')
plt.title("XGBoost  Top 20 Feature Importances")
plt.show()

# Logistic Regression coefficients (absolute values)
lr = LogisticRegression(max_iter=2000)
lr.fit(X_train, y_train)
lr_coef = pd.Series(np.abs(lr.coef_[0]), index=X_train_df.columns).sort_values(ascending=False).head(20)
plt.figure(figsize=(7,5))
lr_coef[::-1].plot(kind='barh')
plt.title("Logistic Regression  Top 20 |coef|")
plt.show()
```
